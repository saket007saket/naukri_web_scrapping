```python
jobs={"link":[]}
```

```python
def get_url(position):
  template ="https://www.naukri.com/"
  pos=''
  #position='Data Scientist'
  p=position.lower()
  p=p.split()
  for i in p:
    pos= pos+i+'-'
  pos=pos+'jobs'+'-'+'in'

  template=template+pos+'?'+'k={}'
  #url=template
  url=template.format(position.lower())
  return url
```

```python
url = get_url('machine learning')
```

```python
while True:
    driver.get(url)
    time.sleep(3)
    lst=driver.find_elements_by_css_selector(".jobTuple.bgWhite.br4.mb-8")
    for job in lst:
        driver.implicitly_wait(10)
        link=job.find_element_by_css_selector('a.title.fw500.ellipsis').get_attribute('href')
        jobs["link"].append(link)
    try:
        driver.find_element_by_css_selector('a.fright.fs14.btn-secondary.br2').click()
    except:
        break
```

```python

while True:
    driver.get(url)
    time.sleep(3)
    lst=driver.find_elements_by_css_selector(".jobTuple.bgWhite.br4.mb-8")
    for job in lst:
        driver.implicitly_wait(10)
        role=job.find_element_by_css_selector("a.title.fw500.ellipsis").text
        company_name=job.find_element_by_css_selector("a.subTitle.ellipsis.fleft").text
        location_of_job=job.find_element_by_css_selector(".fleft.grey-text.br2.placeHolderLi.location").text
        exp=job.find_element_by_css_selector(".ellipsis.fleft.fs12.lh16").text
        skills=job.find_element_by_css_selector(".tags.has-description").text
        salary=job.find_element_by_css_selector(".fleft.grey-text.br2.placeHolderLi.salary").text
        job_desc=job.find_element_by_css_selector(".job-description.fs12.grey-text").text
        jobs["Roles_for_applicant"].append(role)
        jobs["Companies_name"].append(company_name)
        jobs["Location_of_job"].append(location_of_job)
        jobs["Experience"].append(exp)
        jobs["Skills_set_required"].append(skills)
        jobs["Salary_offered"].append(salary)
        jobs["Description_about_job"].append(job_desc)
    try:
        driver.find_element_by_css_selector('a.fright.fs14.btn-secondary.br2').click()
    except:
        break
```
